```python
import pandas as pd
```

```python
df1 = pd.read_csv('g1_modelB_pairs2.csv') # valor semantico de la palabra (3 modelos * 50 expansiones * 2 grupos de consultas)
df2 = pd.read_csv('g2_modelB_pairs2.csv') # junto a diversidad obtenida
```

```python
df1
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots(figsize=(10,6))

sns.scatterplot(ax=ax, data=df1, x="similarity_w", y="sum_cos", s=65, color='#009f75', alpha=0.85)
sns.scatterplot(ax=ax, data=df2, x="similarity_w", y="sum_cos", color='#ef4444', s=65, alpha=0.85)
plt.axhline(y=2.845, color ='g', label='2.845')
plt.axhline(y=2.905, color ='r', label='2.905')
plt.legend(labels=['2.845 (G1)', '2.905 (G2)', 'Group 1', 'Group 2'])
bottom, top = plt.ylim()
plt.ylabel("sum_cos")
#print(bottom, top)
plt.ylim(bottom=1.85, top= 3.4)
```
