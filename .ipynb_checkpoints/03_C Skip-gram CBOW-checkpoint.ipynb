**Word embedding:** para cuantificar la semantica, poder captar el significado
Se quiere una representación que permita corelación entre palabras que tienen un mismo significado o uso

**Continuous Bag of Words Model (CBOW) and Skip-gram** ambas son arquitecturas para aprender las representaciones de palabras subyacentes para cada palabra mediante el uso de redes neuronales

En el modelo CBOW, las representaciones distribuidas de contexto (o palabras circundantes) se combinan para predecir la palabra en el medio.

Mientras que en el modelo Skip-gram, la representación distribuida de la palabra de entrada se usa para predecir el contexto.

```python
import bs4 as bs
import urllib.request
import re
import nltk

scrapped_data = urllib.request.urlopen('https://en.wikipedia.org/wiki/Artificial_intelligence')
article = scrapped_data .read()

parsed_article = bs.BeautifulSoup(article,'lxml')

paragraphs = parsed_article.find_all('p')

article_text = ""

for p in paragraphs:
    article_text += p.text
```

```python
nltk.download('punkt')
```

```python
# Cleaing the text
processed_article = article_text.lower()
processed_article = re.sub('[^a-zA-Z]', ' ', processed_article )
processed_article = re.sub(r'\s+', ' ', processed_article)

# Preparing the dataset
all_sentences = nltk.sent_tokenize(processed_article)

all_words = [nltk.word_tokenize(sent) for sent in all_sentences]

# Removing Stop Words
from nltk.corpus import stopwords
for i in range(len(all_words)):
    all_words[i] = [w for w in all_words[i] if w not in stopwords.words('english')]
```

```python
from gensim.models import Word2Vec #creación del modelo

word2vec = Word2Vec(all_words, min_count=2) #incluir las palabras que solo aparecen al menos dos veces en el corpus
```

```python
# para ver el diccionario
vocabulary = word2vec.wv.vocab
print(vocabulary)
```
