### Preprocesamiento del corpus

Similar a lo realizado en 01_medicion_de_diversidad. Se genera un string auxiliar para lidiar con el manejo de tildes que permite unidecode.

```python
import pandas as pd

corpus = df[:48].str.cat() # concatenee 48 snippets de la primera query del usuario
```

```python
#https://pynative.com/python-generate-random-string/

import string
from random import choice

letters = string.ascii_lowercase
aux_str = ''.join(choice(letters) for i in range(10))
aux_str # se debe verificar que sea una palabra unica al menos en teoria
```

```python
from preprocessing import preprocessing # funcion propia
```

```python
preprocessed_corpus = preprocessing(aux_str, corpus) # no tildes, ni signos ni espacios adicionales, e incluye 'ñ'
```

```python
import nltk

# Preparación del dataset
all_sentences = nltk.sent_tokenize(preprocessed_corpus)

all_words = [nltk.word_tokenize(sent) for sent in all_sentences]
```

```python
from nltk.corpus import stopwords

es_sw = stopwords.words('spanish') # lista stopword a usar
```

```python
# Eliminando stop words
for i in range(len(all_words)):
    all_words[i] = [w for w in all_words[i] if w not in es_sw]
```

### Creación del modelo
Utilizando Gensim basado en Skip Gram

```python
from gensim.models import Word2Vec
```

```python
model = Word2Vec(all_words, min_count=4, sg=1, window=5)
```

```python
for i, word in enumerate(model.wv.vocab):
    if i == 9:
        break
    print(word)
```

```python
model.wv['auto']
```

```python
try:
    display(model.wv.most_similar('auto', topn=30))
except KeyError:
    print("Esta palabra no aparece en el modelo")
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
# por último podemos graficar el modelo
vocab = list(model.wv.vocab)
X = model[vocab]

# aplicamos TSNE
tsne = TSNE(n_components=2) #herramienta para visualizar data de alta dimensionalidad
X_tsne = tsne.fit_transform(X)

# transformamos en DataFrame
df = pd.concat([pd.DataFrame(X_tsne),
                pd.Series(vocab)],
               axis=1)

df.columns = ['x', 'y', 'palabra']
```

```python
len(vocab)
```

```python
# creamos el gráfico
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1)

# reducimos el dataset a 15 palabras para el ejemplo
df_e = df.head(len(vocab))

ax.scatter(df_e['x'], df_e['y'])

for i, txt in enumerate(df_e['palabra']):
    ax.annotate(txt, (df_e['x'].iloc[i], df_e['y'].iloc[i]))
    
plt.show()
```
