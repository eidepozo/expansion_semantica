https://pypi.org/project/Unidecode/

unidecode() takes Unicode data and tries to represent it in ASCII characters where the compromises taken when mapping between two character sets are chosen to be near what a human with a US keyboard would choose.

Strings with accented or special characters are unicode strings while regular one’s ascii.

```python
unaccented_snippets = unidecode(corpus) # como lidiar con tildes sin afectar ñ
# deberia eliminar palabras en ingles?
```

```python
processed_snippets = unaccented_snippets.lower()
processed_snippets = re.sub('[^a-zA-Z]', ' ', processed_snippets)
processed_snippets = re.sub(r'\s+', ' ', processed_snippets)
```

```python
# Preparing the dataset
all_sentences = nltk.sent_tokenize(processed_snippets)

all_words = [nltk.word_tokenize(sent) for sent in all_sentences]

# Removing Stop Words
from nltk.corpus import stopwords
for i in range(len(all_words)):
    all_words[i] = [w for w in all_words[i] if w not in es_sw]
```

```python
#all_words
```

```python
from gensim.models import Word2Vec
```

```python
model = Word2Vec(all_words, min_count=4, sg=1, window=5) #con el cambio 
```

```python
for i, word in enumerate(model.wv.vocab):
    if i == 9:
        break
    print(word)
```

```python
try:
    display(model.wv.most_similar('auto', topn=30))
except KeyError:
    print("The word does not appear in this model")
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
# por último podemos graficar el modelo
vocab = list(model.wv.vocab)
X = model[vocab]

# aplicamos TSNE
tsne = TSNE(n_components=2) #herramienta para visualizar data de alta dimensionalidad
X_tsne = tsne.fit_transform(X)

# transformamos en DataFrame
df = pd.concat([pd.DataFrame(X_tsne),
                pd.Series(vocab)],
               axis=1)

df.columns = ['x', 'y', 'palabra']
```

```python
len(vocab)
```

```python
# creamos el gráfico
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1)

# reducimos el dataset a 15 palabras para el ejemplo
df_e = df.head(len(vocab))

ax.scatter(df_e['x'], df_e['y'])

for i, txt in enumerate(df_e['palabra']):
    ax.annotate(txt, (df_e['x'].iloc[i], df_e['y'].iloc[i]))
    
plt.show()
```

```python
import spacy
import textacy
# Cargando el modelo en español de spacy
#!python -m spacy download es
nlp = textacy.load_spacy_lang('es')
```

```python
nlp
```

```python
corpus_procesado = nlp(corpus)
```

```python
corpus_procesado
```

```python
sentencias = [s for s in corpus_procesado.sents]
```

```python
stop = list(spacy.lang.es.STOP_WORDS) # cambio en comparación a ej
stop[:15]
```

```python
import gensim
```

```python
# transformando el texto para pasarlo al modelo de gensim
texto = [[str(palabra).lower() for palabra in sent if str(palabra) not in stop ]
         for sent in sentencias]
# generando el diccionario
diccionario = gensim.corpora.Dictionary(texto)

```

```python
# creando el modelo
modelo = gensim.models.Word2Vec(texto, workers=4, size=100, 
                                min_count=5, window=10, sample=1e-3)
```

```python
for i, word in enumerate(modelo.wv.vocab): #no elimino simbolos
    if i == 100:
        break
    print(word)
```

```python
# representación de la palabra hermano como vector.
modelo.most_similar_cosmul('auto')

```
