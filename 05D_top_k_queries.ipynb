```python
def guided_df(df):
    a_df = pd.DataFrame(columns = ['snippet1', 'snippet2', 'snippet3'])
    q_set = df['query'].tolist()
    for query in q_set:
        res = bing_search_alt(query)
        a_series = pd.Series(res, index=a_df.columns)
        a_df = a_df.append(a_series, ignore_index=True)
    newdf = pd.concat([df, a_df], axis=1)
    return newdf     
```

```python
df2 = guided_df(df)
```
