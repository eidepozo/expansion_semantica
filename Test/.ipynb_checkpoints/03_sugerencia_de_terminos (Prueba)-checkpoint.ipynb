### Formulación y preprocesamiento del corpus

Similar a lo realizado en 01_medicion_de_diversidad. Se genera un string auxiliar para lidiar con el manejo de tildes que permite unidecode.

```python
import pandas as pd

corpus = df.str.cat() # concatenee los snippets de la primera query del usuario

#https://rare-technologies.com/word2vec-tutorial/
#gensim’s word2vec expects a sequence of sentences
#sentences = [['first', 'sentence'], ['second', 'sentence']]
```

```python
#https://pynative.com/python-generate-random-string/

import string
from random import choice

letters = string.ascii_lowercase
aux_str = ''.join(choice(letters) for i in range(10))
aux_str # se debe verificar que sea una palabra unica al menos en teoria para reemplazo 'ñ'
```

```python
from preprocessing import preprocessing # funcion propia
```

```python
preprocessed_corpus = preprocessing(aux_str, corpus) # no tildes, ni signos ni espacios adicionales, e incluye 'ñ'
```

```python
import nltk

# Preparación del dataset
all_sentences = nltk.sent_tokenize(preprocessed_corpus)

all_words = [nltk.word_tokenize(sent) for sent in all_sentences] #lista dentro de una lista
```

```python
all_words
```

```python
len(all_words[0])
```

```python
from nltk.corpus import stopwords

es_sw = stopwords.words('spanish') # lista stopword a usar
es_sw.extend(['si', 'asi', 'ser', 'tener', 'mas']) # no estaban incluidas antes y se repiten bastante

en_sw = stopwords.words('english') # se incluyen ahora las sw en ingles
```

```python
es_sw
```

```python
# Eliminando stop words
for i in range(len(all_words)):
    all_words[i] = [w for w in all_words[i] if w not in es_sw and w not in en_sw]
```

```python
len(all_words[0])
```

### Creación del modelo
Utilizando Gensim basado en Skip Gram

**Ajustes en el modelo, se evalua cualitativamente min_count**

```python
from gensim.models import Word2Vec
```

```python
model = Word2Vec(all_words, min_count=40, sg=1, window=5) #frecuencia permitida menor que, distancia entre palabras para predicción
#https://radimrehurek.com/gensim/models/word2vec.html#gensim.models.word2vec.Word2Vec
```

```python
vocab = list(model.wv.vocab)
display(len(vocab), vocab)
```

```python
for i, word in enumerate(model.wv.vocab):
    if i == 9:
        break
    print(word)
```

```python
model.wv['auto']
```

```python
model.most_similar(positive=['construir'])
```

```python
try:
    display(model.wv.most_similar('auto', topn=30))
except KeyError:
    print("Esta palabra no aparece en el modelo")
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
# por último podemos graficar el modelo
vocab = list(model.wv.vocab)
X = model[vocab]

# aplicamos TSNE
tsne = TSNE(n_components=2) #herramienta para visualizar data de alta dimensionalidad
X_tsne = tsne.fit_transform(X)

# transformamos en DataFrame
df = pd.concat([pd.DataFrame(X_tsne),
                pd.Series(vocab)],
               axis=1)

df.columns = ['x', 'y', 'palabra']
```

```python
len(vocab)
```

```python
# creamos el gráfico
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1)

# reducimos el dataset a 15 palabras para el ejemplo
df_e = df.head(len(vocab))

ax.scatter(df_e['x'], df_e['y'])

for i, txt in enumerate(df_e['palabra']):
    ax.annotate(txt, (df_e['x'].iloc[i], df_e['y'].iloc[i]))
    
plt.show()
```
